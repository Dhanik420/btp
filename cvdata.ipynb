```
colums=[]
colums.append('K')
for temp in range(4):      #preparing headings for df
    colums.append('B'+str(temp))
    colums.append('T'+str(temp))
    colums.append('H'+str(temp))
for temp in range(2500):      #preparing headings for df
    colums.append('Ts'+str(temp))
df=pd.DataFrame(columns=colums)
'''-----------------initilizing---------------------'''
hc=np.array([0,0,0,0])
bc=np.array([0,0,0,0])
Tc=np.array([0,0,0,0])
b=np.array([0,0,0,0])
itter=0
print('hi')
for bc[0] in range(3):
  for i100 in range(10):
    Tc[0]=(np.random.uniform()*100*10**3 if bc[0]==2 else np.random.uniform()*1000)
    hc[0]=((np.random.uniform()*5+1)*10**3 if bc[0]==1 else 0)
    K=np.random.uniform()*400+50
    Tss=np.zeros((50,50))
    Tss2=np.zeros((50,50))
    trr=0.001
    err=1.
    while err>=trr:
      for i in range(50):
        for j in range(50):
          b[0]=int(i==0)
          b[2]=int(i==49)
          b[1]=int(j==49)
          b[3]=int(j==0)
          if sum(b)>0:
            isoth=np.where(((b==1)*(bc==0))==1)
            if np.size(isoth):
              Tss2[i,j]=Tc[isoth[0][0]]
              continue
            con=b[0]*((bc[0]==1)*(Tc[0]*hc[0]/50)+(bc[0]==2)*(Tc[0]/50))+(0 if b[0] else (K*Tss2[i-1,j]))+b[1]*((bc[1]==1)*(Tc[1]*hc[1]/50)+(bc[1]==2)*(Tc[1]/50))+(0 if b[1] else (K*Tss2[i,j+1]))+b[2]*((bc[2]==1)*(Tc[2]*hc[2]/50)+(bc[2]==2)*(Tc[2]/50))+(0 if b[2] else (K*Tss2[i+1,j]))+b[3]*((bc[3]==1)*(Tc[3]*hc[3]/50)+(bc[3]==2)*(Tc[3]/50))+(0 if b[3] else (K*Tss2[i,j-1]))
            mul=b[0]*((bc[0]==1)*(hc[0]/50))+(not b[0])*K+b[1]*((bc[1]==1)*(hc[1]/50))+(not b[1])*K+b[2]*((bc[2]==1)*(hc[2]/50))+(not b[2])*K+b[3]*((bc[3]==1)*(hc[3]/50))+(not b[3])*K
            Tss2[i,j]=con/mul
          else :
            Tss2[i,j]=(Tss2[i-1,j]+Tss2[i+1,j]+Tss2[i,j-1]+Tss2[i,j+1])/4
      err=abs(Tss-Tss2).max()
      Tss=Tss2.copy()
      '''-----------------------------CFD code------------------------------------'''
    Tss=Tss.reshape(2500)                         #reshape result 2d to 1d arr
    colval=[]
    colval.append(K)
    for temp in range(4):
      colval.append(bc[temp])
      colval.append(Tc[temp])
      colval.append(hc[temp])
    colval=np.array(colval)
    colval=np.concatenate((colval,Tss))
    df=df.append(pd.DataFrame([colval],columns=colums)) #apppend result and features
    print(itter*100/30)
    itter+=1
```
